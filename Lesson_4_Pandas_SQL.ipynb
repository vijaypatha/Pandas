```python
import pandas as pd 
import numpy as np
```

```python
url = ('https://raw.github.com/pandas-dev'
         '/pandas/master/pandas/tests/data/tips.csv')
```

```python
tips = pd.read_csv(url)
tips.head(4)
```

## Selecting

```python
'''
Select total_bill, sex, day
From tips
limit 5
'''

tips[["total_bill","sex", "day"]].head(5)
```

## Where 

```python
tips[tips["time"] == "Dinner"].head(5)
```

## Conditions 

```python
'''
-- tips of more than $5.00 at Dinner meals
SELECT *
FROM tips
WHERE time = 'Dinner' AND tip > 5.00;
'''

tips[(tips['time'] == 'Dinner') & (tips['tip'] > 5)].head(5)
```

```python
'''
-- tips by parties of at least 5 diners OR bill total was more than $45
SELECT *
FROM tips
WHERE size >= 5 OR total_bill > 45;
'''
```

```python
tips[(tips["size"] >= 5) | (tips["total_bill"]>45)]
```

```python
tips[tips.isna().any(axis=1)]
```

## Group BY

```python
'''
SELECT sex, count(*)
FROM tips
GROUP BY sex;
/*
Female     87
Male      157
*/
'''
```

```python
tips.groupby('sex').size()
```

```python
# count() applies the function to each column, returning the number of not null records within each.
tips.groupby('sex').count()
```

```python
'''
SELECT day, AVG(tip), COUNT(*)
FROM tips
GROUP BY day;
/*
Fri   2.734737   19
Sat   2.993103   87
Sun   3.255132   76
Thur  2.771452   62
*/
'''
```

```python
tips.groupby('day').agg({'tip':np.mean, 'day':np.size})
```
